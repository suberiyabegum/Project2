```python
# Numerical columns → MEDIAN
num_cols = [
    'Customer_Age', 'Delivery_Time_Min',
    'Distance_km', 'Order_Value', 'Delivery_Rating'
]
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns → 'Others'
cat_cols = ['Customer_Gender', 'City', 'Area', 'Cuisine_Type']
df[cat_cols] = df[cat_cols].fillna('Others')

# Payment mode → MODE
df['Payment_Mode'].fillna(df['Payment_Mode'].mode()[0], inplace=True)


```

```python
df[['Delivery_Time_Min', 'Order_Value']].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure()
df.boxplot(column = 'Order_Value')
plt.title("Before Outlier Capping - Order_Value")
plt.show()
```

# capping outliers

```python
# Delivery time bounds
lower_dt = df['Delivery_Time_Min'].quantile(0.01)
upper_dt = df['Delivery_Time_Min'].quantile(0.95)

df['Delivery_Time_Min'] = df['Delivery_Time_Min'].clip(lower_dt, upper_dt)

# Order value bounds
lower_ov = df['Order_Value'].quantile(0.01)
upper_ov = df['Order_Value'].quantile(0.95)

df['Order_Value'] = df['Order_Value'].clip(lower_ov, upper_ov)
```

# correcting invalid values

```python
# Fix invalid ratings
df.loc[df['Delivery_Rating'] > 5, 'Delivery_Rating'] = 5
df.loc[df['Delivery_Rating'] < 1, 'Delivery_Rating'] = 1

# Fix negative profit margin
df.loc[df['Profit_Margin'] < 0, 'Profit_Margin'] = 0
```

# standardising categorical values

```python
cat_columns = [
    'Customer_Gender', 'City', 'Area',
    'Cuisine_Type', 'Payment_Mode', 'Order_Status'
]

for col in cat_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
```

# Ensuring logical consistency 

```python
# cancellation_reason
df.loc[
    (df['Order_Status'] == 'Delivered') & (df['Cancellation_Reason'].isna()),
    'Cancellation_Reason'
] = 'Not Cancelled'


# cancelled orders with missing reason
df.loc[
    (df['Order_Status'] == 'Cancelled') & (df['Cancellation_Reason'].isna()),
    'Cancellation_Reason'
] = 'Unknown'


# cancelled orders should not have final_amount
df.loc[
    df['Order_Status'] == 'Cancelled',
    'Final_Amount'
] = 0
```
